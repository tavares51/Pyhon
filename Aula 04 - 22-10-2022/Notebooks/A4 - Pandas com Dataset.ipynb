```python
# Criando uma nova coluna e aplicando uma operaÃ§Ã£o matemÃ¡tica entre as variÃ¡veis:
df['Ticket'] = df['Total'] / df['Quantity']
df.head()
```

### 4 - Utilizando funÃ§Ãµes agregadas:

```python
# Obtendo a mÃ©dia de valores de uma coluna:
df['Ticket'].mean()
```

```python
# Obtendo o valor mÃ­nimo de uma coluna:
df['Total'].min()
```

```python
# Obtendo o valor mÃ¡ximo de uma coluna:
df['Total'].max()
```

```python
# Retornando valores no quantil* fornecido sobre o eixo solicitado:
df.quantile(.25)
```

```python
# Obtendo a moda de alguma coluna (valor que mais aparece):
df['Quantity'].mode()
```

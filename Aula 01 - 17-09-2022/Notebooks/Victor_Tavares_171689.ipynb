1 - Escreva um algoritmo que leia um número inteiro qualquer do usuário, e informe se o mesmo é par ou ímpar. Caso o usuário informe um número real, informar a mensagem "O número não é inteiro.". Caso seja informado um número negativo, informar a mensagem "O número deve ser positivo.".

```python
#Recebendo número do usuário
numero = float(input("Digite um numero: "))

#Verificando se é par ou impar
if numero % 2 == 0:
    print(f"Número {numero} é par")
elif numero % 2 == 1:
    print(f"Número {numero} é impar")

#Verificando se é um numero real
if numero // 1 != numero: 
    print("Número não é inteiro")

#Verificando se é negativo
if numero < 0:
    print("O número deve ser positivo...")
```

2 - Escreva um algoritmo que leia três valores inteiros e diferentes, mostrando-os no final na ordem decrescente. Caso existam valores iguais, informar a mensagem "Foram informados números repetidos.".

```python
#Recebendo dados do usuário
primeiro = int(input("Digite o primeiro número: "))
segundo = int(input("Digite o segundo número: "))
terceiro = int(input("Digite o terceiro número: "))
aux = 0

#Verificar se tem número igual caso contrario analisar do maior para o menor
if (primeiro == segundo) or (primeiro == terceiro) or (segundo == terceiro):
    print("Foram informados números repetidos...")
else:
    if(terceiro > segundo):
        aux = terceiro
        terceiro = segundo
        segundo = aux

    if(segundo > primeiro):
        aux = segundo
        segundo = primeiro
        primeiro = aux

    if(terceiro > segundo):
        aux = terceiro
        terceiro = segundo
        segundo = aux
        
    print(f"{primeiro}, {segundo}, {terceiro}")
```

3 - Escreva um algoritmo que calcule o valor a ser pago por um determinado produto, considerando as seguintes condições:

* A: à vista com 10% de desconto;
* B: em duas vezes, sem juros;
* C: em três vezes, com 5% de juros (acima do valor original).


Para isso, devem ser consideradas duas entradas do usuário: valor do produto e condição de pagamento (A, B ou C). Como saída, deverá ser informado o valor a ser pago com os reajustes (no caso de parcelas, informar o valor de cada parcela).



```python
#Recebendo dados do usuário
valor = float(input("Digite o valor do produto: "))
condicao = input("Digite a condição de pagamento (A, B, C): ")

#Veriricar condição A e aplicar 10% de desconco
if condicao.upper() == 'A':
    desconto = valor * (0.10)
    valor_descontado = valor - desconto
    print(f"Valor a ser pago com 10% de desconto: R$ {valor_descontado}")

#Veriricar condição B e dividir valor em duas vezes
elif condicao.upper() == 'B':
    valor_parcela = valor / 2
    print(f"Valor do produto em 2 vezes, sem juros: R$ {valor_parcela} (por parcela)")

#Veriricar condição C, aidicionar o juros de 5% e dividir por 3 parcelas
elif condicao.upper() == 'C':
    valor_juros = valor + (valor * (0.05))
    valor_parcela = valor_juros / 3
    print(f"Valor do produto em 3 vezes, com juros de 5%: R$ {valor_parcela} (por parcela)")
```

4. Crie um algoritmo que receba o nome e a nota de um aluno, imprimindo no final o seguinte conteúdo: Olá **nome**, você está **situação**.

    ○ Considere que a situação deverá ser atribuída da seguinte forma:

    ■ Nota >= 0 e < 4: REPROVADO;

    ■ Nota >= 4 e < 7: RECUPERAÇÃO;

    ■ Nota >= 7 e <= 10: APROVADO;

    ■ Quaisquer outros valores diferentes desses intervalos, informar a mensagem: Nota informada inválida!

```python
#Obtendo dados do aluno
nome = input("Digite nome do aluno: ")
nota = int(input("Digite a nota do aluno: "))
situacao = ""

#Validando nota do aluno
if nota >= 0 and nota < 4:
    situacao = "REPROVADO"
elif nota >= 4 and nota < 7:
    situacao = "RECUPERAÇÃO"
elif nota >= 7 and nota <= 10:
    situacao = "APROVADO"
else:
    situacao = "Nota informada inválida!"

#Imprimindo Resultado
print(f"Olá {nome}, você está {situacao}")
```

5. Crie um algoritmo que receba os seguintes dados de um determinado
funcionário:

    ○ Salário Bruto

    ○ Tempo de Trabalho (em anos)

    ○ Se utiliza Vale Transporte (S ou N)
    
    Após, utilize os critérios a seguir para calcular o seu salário líquido final, imprimindo
    a seguinte mensagem: O seu salário líquido calculado foi de R$ **valor**.
    
    ● Desconto de INSS:

        ○ Até R$ 1.200,00: 7%

        ○ De R$ 1.200,01 até R$ 5.000,00: 13%

        ○ Acima de R$ 5.000,00: 16%

    ● Desconto de VT (caso seja utilizado): 6%

    ● Aumento (percentual) por tempo de empresa:

        ○ 1 ano: 2%

        ○ De 2 anos até 4 anos: 5%

        ○ De 5 anos até 10 anos: 10%

        ○ Acima de 10 anos (eita): 20%

```python
#Obtendo dados do funcionário
salario_bruto = float(input("Digite o salário do funcionário: "))
tempo_trabalho = int(input("Digite o tempo de trabalho do funcionário: "))
vale_transporte = input("Informe se o funcionário possui vale transporte (S ou N): ")

#Calculando salário liquido
salario_liquido = round(salario_bruto, 2)

#INSS
if salario_bruto <= 1200.00:
    salario_liquido -= (salario_bruto * 0.7)
elif salario_bruto > 1200.00 and salario_bruto <= 5000.00:
    salario_liquido -= (salario_bruto * 0.13)
elif salario_bruto > 5000.00:
    salario_liquido -= (salario_bruto * 0.16)

#Desconto VT
if vale_transporte == 'S':
    salario_liquido -= (salario_bruto * 0.06)

#Aumento
if tempo_trabalho == 1:
    salario_liquido += (salario_bruto * 0.02)
elif tempo_trabalho >= 2 and tempo_trabalho <= 4:
    salario_liquido += (salario_bruto * 0.05)
elif tempo_trabalho >= 5 and tempo_trabalho <= 10:
    salario_liquido += (salario_bruto * 0.10)
elif tempo_trabalho > 10:
    salario_liquido += (salario_bruto * 0.20)

print(f"O seu salário líquido calculado foi de R$ {round(salario_liquido, 2)}")
```

6. Crie um algoritmo que receba o nome e o sobrenome de uma pessoa
(independentemente de letras maiúsculas ou minúsculas), retornando as
seguintes informações:

    ○ O nome e o sobrenome informados possuem a mesma quantidade de
    letras?

    ○ O nome e o sobrenome são iguais?

    ○ Quem possui mais letras (o nome ou o sobrenome)?

    ○ O e-mail dessa pessoa será: **nome**.**sobrenome**@facens.br

    Obs.: lembre-se que e-mails devem possuir apenas letras minúsculas!

```python
#Obtendo dados da pessoa
nome = input("Digite o nome da pessoa: ")
sobrenome = input("Digite o sobrenome da pessoa: ")

#O nome e o sobrenome informados possuem a mesma quantidade de letras?
if len(nome) == len(sobrenome):
    print("O nome e o sobrenome informados possuem a mesma quantidade de letras.")
else:
    print("O nome e o sobrenome informados não possuem a mesma quantidade de letras.")

#O nome e o sobrenome são iguais?
if nome == sobrenome:
    print("O nome e o sobrenome são iguais.")
else:
    print("O nome e o sobrenome não são iguais.")

#Quem possui mais letras (o nome ou o sobrenome)?
if len(nome) > len(sobrenome): 
    print("o nome possui mais letras que o sobrenome.")
elif len(nome) < len(sobrenome): 
    print("o sobrenome possui mais letras que o nome.")

#email
print("Email", f"{nome.lower()}.{sobrenome.lower()}@facens.br")
```

7. Crie um algoritmo que resolva a Fórmula de Bhaskara (tendo como entrada do
usuário valores para A, B e C).

```python
#Obtendo A, B e C
a = float(input("Digite A:"))
b = float(input("Digite B:"))
c = float(input("Digite C:"))

x1 = ((-1 * b) + pow((b ** 2 - (4*a*c)), 0.5)) / 2*a
x2 = ((-1 * b) - pow((b ** 2 - (4*a*c)), 0.5)) / 2*a

print("x1", x1)
print("x2", x2)
```

8. Crie um algoritmo que receba os três lados de um triângulo, e classifique-o
como: equilátero, isósceles ou escaleno.

```python
#Obtendo A, B e C
lado_um = float(input("Lado 1: "))
lado_dois = float(input("Lado 2: "))
lado_tres = float(input("Lado 3: "))

if lado_um == lado_dois == lado_tres:
    print("equilatero")
elif (lado_um == lado_dois) or (lado_um == lado_tres) or (lado_dois == lado_tres):
    print("isóceles")
else:
    print("escaleno")
```

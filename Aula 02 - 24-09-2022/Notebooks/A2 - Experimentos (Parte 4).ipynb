# Lambda Functions:
(Funções básicas, simples...)

```python
# Exemplo de uma função que faz a soma de dois números inteiros:
def sum(n1, n2):
  return n1 + n2

sum(1, 2)
```

```python
# Agora, realizando o mesmo processo com Lambda:
sum = lambda n1, n2: n1 + n2

sum(1, 2)
```

```python
# Mais um exemplo: 
result = lambda x: 'PAR' if x % 2 == 0 else 'ÍMPAR'

result(2)
```

```python
# E mais um...
example_list = [ 1, 2, 3, 4, 5]
result = lambda x: [ y for y in x if y % 2 == 0 ]

result(example_list)
```

```python
# O último...
# Descobrindo qual o maior número (dentre 3 números recebidos):
find_max = lambda a, b, c: a if a > b and a > c else (b if b > a and b > c else c)

find_max(2, 15, 7)
```

```python
soma = lambda lista: [l for l in lista]
```

# Map
(aplica uma função para todos os elementos de um objeto - retornando None caso não seja possível aplicar uma determinada estratégia)

```python
# Exemplo 01:
nums = [ 1, 2, 3, 4, 5 ]  

def mult(n):    
  return n * n

result = list(map(mult, nums))
print(result)
```

```python
# Exemplo 02:
nums = [ 1, 2, 3, 4, 5 ]  

def compare(n):    
  if n % 2 == 0:
    return n 

result = list(map(compare, nums))
print(result)
```

```python
# Exemplo 03:
nums = [ 1, 2, 3, 4, 5 ]

result = list(map(lambda x: x * x, nums))
print(result)
```

# Filter
(retorna um conjunto de dados a partir de alguma função)

```python
seq = [ 0, 1, 2, 3, 4, 5 ]

# Retornando números ímpares:
result = filter(lambda x: x % 8, seq)
print(list(result))

# Retornando números pares:
result = filter(lambda x: x % 2 == 0, seq)
print(list(result))
```

# Zip
(combinação)

```python
names = [ 'Adriano', 'Juliana', 'Carlos' ]
ages = [ 30, 23, 12 ]

result = dict(zip(names, ages))
print(result)


result = list(zip(names, ages))
print(result)
```

```python
# Exercício 01:
import pprint

numbers = [6, 8, 9, 7, 1, 3]

evens = []
odds = []

evens.append(numbers[0]) if numbers[0] % 2 == 0 else odds.append(numbers[0])
evens.append(numbers[1]) if numbers[1] % 2 == 0 else odds.append(numbers[1])
evens.append(numbers[2]) if numbers[2] % 2 == 0 else odds.append(numbers[2])
evens.append(numbers[3]) if numbers[3] % 2 == 0 else odds.append(numbers[3])
evens.append(numbers[4]) if numbers[4] % 2 == 0 else odds.append(numbers[4])
evens.append(numbers[5]) if numbers[5] % 2 == 0 else odds.append(numbers[5])

t = (evens, odds)

pprint.pprint(t)
```

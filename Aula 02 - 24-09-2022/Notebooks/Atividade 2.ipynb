Crie uma calculadora utilizando a Linguagem Python, apresentando o seguinte menu interativo para o usuário:

* A - Somar
* B - Subtrair
* C - Multiplicar
* D - Dividir
* E - Potência
* F - Raiz Quadrada
* G - Média
* H - Máximo
* I - Mínimo
* J - Área
* K - Perímetro
* S - Sair


**Considerações importantes sobre a interação com o usuário:**

1 - O menu deverá ser apresentado enquanto o usuário não informar o comando de saída (no caso, representado pela letra 'S');

2 -  Ao término de cada operação escolhida, a tela deverá ser limpa e apresentado novamente o menu principal;

3 -  Caso o usuário informe alguma opção inválida, deverá possibilitar a inserção novamente (até acertar);

4 -  Para as opções 'A', 'B', 'C', 'D', 'G' e 'I' o usuário poderá passar quantos números desejar (caso informe na sequência dos números algum caracter, solicitar a correção ou voltar ao menu principal - disponibilizar essas duas opções, não esquecer de limpar a tela);

5 -  Para a opção 'E', solicitar a base e o expoente. Caso seja informado algum valor indevido, segue mesmas validações do item 4;

6 - Para a opção 'F', idem correções do item 4 (caso seja informado um valor indevido);

7 - Pars os itens 'J' e 'K', solicitar os dados necessários para os cálculos (escolher uma figura geométrica para cada);


**Considerações importantes sobre o desenvolvimento do código:**

- Procure utilizar com propriedade os conceitos aprendidos em sala de aula;
- Reutilizem o código com o uso de funções;



```python
import math
from functools import reduce

caracteres = 'abcçdefghijklmnopqrstuvwxyzABCÇDEFGHIJKLMNOPQRSTUVWXYZ!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~ \t\n\r\x0b\x0c'
```

```python
# Pega números
def pega_numeros(operacao):
    numeros = [] # Definindo lista de numeros
    finaliza = False # Varíavel para finalizar o laço

    while not finaliza:
        numero = input(f'Informe um número para a operação {operacao}: ')

        if numero in caracteres:
            print("Caractere digitado, por favor digite somente números...")                
        else:
            numeros.append(float(numero))
            questao = input('Deseja inserir mais um número [S], Para finalizar [F]: ')
            if questao.upper() == 'F':
                finaliza = True
    return numeros
```

```python
# Função de SOMA
def somar(numeros):
    return reduce(lambda x,y: x + y, numeros) #retorna a soma de todos os elemento de lista
```

```python
# Função de SUBTRAÇÃO
def subtrair(numeros):
    return reduce(lambda x,y: x - y, numeros) #retorna a subtrair de todos os elemento de lista
```

```python
# Função de SUBTRAÇÃO
def multiplicar(numeros):
    return reduce(lambda x,y: x * y, numeros) #retorna a multiplicar de todos os elemento de lista
```

```python
# Função de DIVISÃO
def dividir(numeros):
    return reduce(lambda x,y: x / y, numeros) #retorna a divisão de todos os elemento de lista
```

```python
# Função de MÉDIA
def media(numeros):
    total = len(numeros)
    soma = 0
    for numero in numeros:
        soma += numero
    return soma / total
```

```python
# Função MAXIMO
def maximo(numeros):
    maior = 0
    for numero in numeros:
        if numero > maior: 
            maior = numero
    return maior
```

```python
# Função MENOR
def menor(numeros):
    menor = numeros[0]
    for numero in numeros:
        if numero < menor: 
            menor = numero
    return menor
```

```python
# Função POTÊNCIA
def potencia():
    finalizar = False
    retorno = 0
    while not finalizar:
        base = input("Digite a base: ")
        expoente = input("Digite o expoente: ")
        
        if base in caracteres or expoente in caracteres:
            print("Caractere digitado por favro insira somente números.")
        else:
            retorno = float(base) ** float(expoente)
            finalizar = True
    return retorno
```

```python
# Função RAIZ QUADRADA
def raiz():
    finalizar = False
    retorno = 0
    while not finalizar:
        numero = input("Digite o valor para raiz quadrada: ")
        
        if numero in caracteres:
            print("Caractere digitado por favro insira somente números.")
        else:
            if float(numero) > 0:
                retorno = math.sqrt(float(numero))
                finalizar = True
            else:
                print("Não é possivel calcular raiz quadrada de número negativo.")
    return retorno
```

```python
# Função PERÍMETRO
def perimetro():
    print("\nCalculo perimetro de um HEXÁGONO")
    perimetro_retorno = 0
    for i in range(6):
        lado = float(input(f"Digite valor do lado {i+1}: "))
        perimetro_retorno += lado
    return perimetro_retorno
```

```python
# Função ÁREA
def area():
    print("\nCalculo área de um Triângulo, passando a base e altura.")
    base = float(input("Digite o valor da Base: "))
    altura = float(input("Digite o valor da Altura: "))
    return (base * altura) / 2
```

```python
# Menu
finalizar = False

print("------- Menu Calculadora -------")
while not finalizar:
    print("\nOpções")
    print("* A - Somar\n* B - Subtrair\n* C - Multiplicar\n* D - Dividir\n* E - Potência\n* F - Raiz Quadrada\n* G - Média\n* H - Máximo\n* I - Mínimo\n* J - Área (Triângulo)\n* K - Perímetro (Hexagono)\n* S - Sair")
    opcao = input("Digite a Opção Desejada: ")
    
    if opcao.upper() == 'A':
        numeros = pega_numeros('Somar')
        resultado_soma = somar(numeros)
        print("Soma:", resultado_soma)

    elif opcao.upper() == 'B':
        numeros = pega_numeros('Subtrair')
        resultado_subtracao = subtrair(numeros)
        print("Subtração:", resultado_subtracao)

    elif opcao.upper() == 'C':
        numeros = pega_numeros('Multiplicar')
        resultado_multiplicacao = multiplicar(numeros)
        print("Multiplicação:", resultado_multiplicacao)

    elif opcao.upper() == 'D': 
        numeros = pega_numeros('Dividir')
        resultado_divisao = dividir(numeros)
        print("Divisão:", resultado_divisao)

    elif opcao.upper() == 'E': 
        resultado_potencia = potencia()
        print("Potência", resultado_potencia)

    elif opcao.upper() == 'F': 
        resultado_raiz = raiz()
        print("Raiz", resultado_raiz)

    elif opcao.upper() == 'G':
        numeros = pega_numeros('Média')
        resultado_media = media(numeros)
        print("Média", resultado_media)

    elif opcao.upper() == 'H':
        numeros = pega_numeros('Máximo') 
        resultado_maximo = maximo()
        print("Máximo", resultado_maximo)

    elif opcao.upper() == 'I': 
        numeros = pega_numeros('Menor') 
        resultado_menor = menor()
        print("Menor", resultado_menor)

    elif opcao.upper() == 'J': 
        resultado_area = area()
        print("Área (Triângulo)", resultado_area)

    elif opcao.upper() == 'K': 
        resultado_perimetro = perimetro()
        print("Perímetro (Hexágono)", resultado_perimetro)

    elif opcao.upper() == 'S':
        finalizar = True
     
```

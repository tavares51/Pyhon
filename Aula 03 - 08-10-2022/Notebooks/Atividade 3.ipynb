Crie uma função que efetue a leitura do arquivo e processe seus dados identificando o total de ocorrencias no período avaliado e, baseado nessa informação criar um arquivo de saída em formato JSON (statistics.json) com as seguintes as estatísticas (número de ocorrências e percentual do total) por fase de operação (coluna aeronave_fase_operacao), como o exemplo a seguir:

Crie um arquivo de saida (formato CSV) com nome levels.csv contendo as seguintes informações:

* operation -> aeronave_operador_categoria
* type -> aeronave_tipo_veiculo
* manufacturer -> aeronave_fabricante
* engine_type aeronave_motor_tipo
* engines -> aeronave_motor_quantidade
* year_manufacturing -> aeronave_ano_fabricacao 
* seating -> aeronave_assentos
* fatalities -> total_fatalidades

Considerando apenas acidentes cujo nível de dano da aeronave tenha sido LEVE ou NENHUM (coluna aeronave_nivel_dano) E que o número de fatalidades (total_fatalidades) tenha sido superior à 0 (zero).

**Observações**: 
1. Efetuar apenas uma leitura do arquivo de entrada
2. Efetuar todas as leituras e gravações de arquivos na pasta /tmp

````python
#importção biblioteca CSV, JSON
import csv
import json
````

````python
def read_anv():
    with open('tmp/anv.csv', 'r', encoding='UTF-8') as arquivo:
        arquivo = csv.DictReader(arquivo, delimiter='~')
        statics = []
        fases = []

        #Levels
        with open("tmp/levels.csv", "w" , newline="") as levels:
            colunas = [
                "operation",
                "type",
                "manufacturer",
                "engine_type",
                "engines",
                "year_manufacturing",
                "seating",
                "fatalities"
            ]

            writer = csv.DictWriter(levels,fieldnames=colunas)
            writer.writeheader()

            for linha in arquivo:
                fases.append(linha['aeronave_fase_operacao'])
                if int(linha['total_fatalidades']) > 0 and (linha['aeronave_nivel_dano'] == 'LEVE' or linha['aeronave_nivel_dano'] == 'NENHUM'):
                    writer.writerow({
                            "operation": linha['aeronave_operador_categoria'],
                            "type": linha['aeronave_tipo_veiculo'],
                            "manufacturer": linha['aeronave_fabricante'],
                            "engine_type": linha['aeronave_motor_tipo'],
                            "engines": linha['aeronave_motor_quantidade'],
                            "year_manufacturing": linha['aeronave_ano_fabricacao'],
                            "seating": linha['aeronave_assentos'],
                            "fatalities": linha['total_fatalidades']
                        })

        #Statics
        fases_label = set(fases)
        total = len(fases)
        for f in fases_label:
            fase_operacao = f
            ocorrencias = len([fase for fase in fases if fase==f])
            percentual = (ocorrencias / total) * 100
            static = {
                    "fase_operacao": fase_operacao, 
                    "ocorrencias": ocorrencias,
                    "percentual": str("%.2f" % percentual).replace('.', ',') + "%"
                }
            statics.append(static)

        with open('tmp/statics.json', 'w', encoding='UTF-8') as statics_file:
            json.dump(statics, statics_file, ensure_ascii=False, indent=4) 
````

````python
read_anv()
````

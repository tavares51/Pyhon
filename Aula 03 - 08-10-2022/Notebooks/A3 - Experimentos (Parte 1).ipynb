## Parte 03:

*   Atributos públicos e privados
*   Encapsulamento
*   Herança simples

```python
class Pessoa():
  def __init__(self, nome, idade):
    self.__nome = nome
    self.__idade = idade
  
  @property
  def nome(self):
    return self.__nome

  @property
  def idade(self):
    return self.__idade

  @nome.setter
  def nome(self, nome):
    self.__nome = nome

  @idade.setter
  def idade(self, idade):
    self.__idade = idade

  def apresentar(self):
    return f'Olá! Meu nome é { self.__nome } e tenho { self.__idade } anos de idade.'


class Aluno(Pessoa):
  def __init__(self, nome, idade):
    super().__init__(nome, idade)

adriano = Aluno('Adriano', 30)

adriano.nome = 'Adriano Silva'
adriano.idade = 29

print(type(adriano))

adriano.apresentar()
```

## Parte 04:

*   Polimorfismo
*   Sobrescrita

```python
class PessoaA():
  def apresentar(self):
    print('Apresentando a pessoa A')

class PessoaB(PessoaA):
  def apresentar(self):
    print('Apresentando a pessoa B')

class PessoaC(PessoaB):
  pass


pessoa_a = PessoaA()
pessoa_a.apresentar()

pessoa_b = PessoaB()
pessoa_b.apresentar()

pessoa_c = PessoaC()
pessoa_c.apresentar()

#####


class PessoaX():
  def __str__(self):
    return 'Apresentando a Pessoa X'

class PessoaY(PessoaX):
  def __str__(self):
    return super().__str__() + ' ops... Pessoa Y'


pessoa_x = PessoaX()
print(pessoa_x)

pessoa_y = PessoaY()
print(pessoa_y)
```

## Parte 05:

*   Herança múltipla

```python
class AnimalTerrestre():
  def __init__(self):
    self.habitat = 'terra'

  def andar(self):
    print('Andando...')

class AnimalAquatico():
  def __init__(self):
    self.habitat = 'agua'
    # self.teste = 'Teste'

  def nadar(self):
    print('Nadando...')

class Anfibio(AnimalAquatico, AnimalTerrestre):
  def __init__(self):
      AnimalTerrestre.__init__(self)
      AnimalAquatico.__init__(self)
      # super().__init__()
    
  def andar(self):
    print('Anfíbio Andando...')
  
  

x = Anfibio()

print(x.habitat)
#print(x.teste)
x.andar()
x.nadar()

```

## Parte 06:

*   Associação por agregação
*   Associação por composição

```python
'''
  Associação (Agregação)
'''

class Produto():
  def __init__(self, nome, valor):
    self.nome = nome
    self.valor = valor


class CarrinhoCompras():
  def __init__(self):
    self.produtos = []

  def inserir_produto(self, produto):
    self.produtos.append(produto)

  def listar_produtos(self):
    for produto in self.produtos:
      print(produto.nome, produto.valor)


produto_1 = Produto('Teclado', 35.99)
produto_2 = Produto('Mouse', 12.95)

carrinho_compras = CarrinhoCompras()

carrinho_compras.inserir_produto(produto_1)
carrinho_compras.inserir_produto(produto_2)

carrinho_compras.listar_produtos()

```

```python
'''
  Associação: Composição
'''

class Cliente():
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade
    self.enderecos = []

  def inserir_endereco(self, cidade, estado):
    self.enderecos.append(Endereco(cidade, estado))

  def listar_enderecos(self):
    for endereco in self.enderecos:
      print(endereco.cidade, endereco.estado)

  def __del__(self):
    print(f'O cliente { self.nome } foi apagado!')

class Endereco():
  def __init__(self, cidade, estado):
    self.cidade = cidade
    self.estado = estado

  def __del__(self):
    print(f'O endereço { self.cidade, self.estado } foi apagado!')


cliente_1 = Cliente('Adriano', 30)
cliente_1.inserir_endereco('Salto', 'SP')
cliente_1.inserir_endereco('Belo Horizonte', 'MG')

print(cliente_1.nome)
cliente_1.listar_enderecos()
print('')

cliente_2 = Cliente('Juliana', 23)
cliente_2.inserir_endereco('São Paulo', 'SP')

print(cliente_2.nome)
cliente_2.listar_enderecos()
print('')

print('-------------- PROGRAMA ENCERRADO!')
```

## Parte 07:

*   Abstração

```python
from abc import ABC, abstractmethod

class Conta(ABC):
  @abstractmethod
  def retirar(self, valor):
    pass

  def depositar(self, valor):
    pass

class ContaCorrente(Conta):
  def __init__(self):
    self.__saldo = 0

  @property
  def saldo(self):
    return self.__saldo

  def retirar(self, valor):
    self.__saldo -= valor

conta_adriano = ContaCorrente()

conta_adriano.retirar(4.0)
conta_adriano.saldo
```

## Parte 08:


*    Sobrecarga de métodos (workaround)

```python
class Operacoes():
  def somar(self, n1 = None, n2 = None):
    resultado = 0

    if n1 != None and n2 != None:
      resultado = n1 + n2
    elif n1 != None and n2 == None:
      resultado = n1
    elif n1 == None and n2 != None:
      resultado = n2
    
    return resultado


operacoes = Operacoes()
print(operacoes.somar(1))
print(operacoes.somar(2, 2))
    
```

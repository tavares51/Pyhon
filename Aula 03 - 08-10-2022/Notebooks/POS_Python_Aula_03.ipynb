```python
class Veiculo():
    # Método Construtor
    def __init__(self):
        # Inicializando atributos
        self.marca = ''
        self.modelo = ''
        self.ano = 0
    
    def mostrar_detalhes(self):
        print(f"Marca: {self.marca}, Modelo: {self.modelo}, Ano: {self.ano}")
```

```python
v = Veiculo()
v.ano = 2006
v.marca = 'VW'
v.modelo = 'Saveiro'
v.mostrar_detalhes()
```

```python
class Residencia():
    def __init__(self, tipo, comodos):
        self.tipo = tipo
        self.comodos = comodos
    
    def mostrar_dados(self):
        print(f"Tipo: {self.tipo}, Cômodos: {self.comodos}")
    
    def __str__(self):
        return f"Tipo: {self.tipo}, Cômodos: {self.comodos}"
```

```python
r = Residencia('Casa', 5)
r.mostrar_dados()
print(r)
```

```python
# Associação por Agregação
class Produto():
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

class CarrinhoDeCompras():
    def __init__(self):
        self.produtos = []
    
    # Quanto parametro do metodo tem referencia a uma possivel classe, isso se chama agregação 
    def inserir_produto(self, produto):
        self.produtos.append(produto)
    
    def listar_produtos(self):
        for produto in self.produtos:
            print(produto.nome, produto.preco)
```

```python
produto_1 = Produto('Farinha', 1.90)
produto_2 = Produto('Sardinha', 6.52)
produto_3 = Produto('Cerveja', 3.50)

carrinho = CarrinhoDeCompras()
carrinho.inserir_produto(produto_1)
carrinho.inserir_produto(produto_2)
carrinho.inserir_produto(produto_3)
carrinho.listar_produtos()
```
